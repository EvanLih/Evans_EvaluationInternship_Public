Importing the uncleaned/internship dataset from Evans Jobs

```python
df = pd.read_excel('Initial_Dataset_through_6.19.2019.xls')

df = df[df['Flags'].str.contains("Entering Class of Fall 2018")]

#df['Agency/Organization Name'].unique()

df.drop(['Student'], axis = 1, inplace = True)

unpaid = len(df[df['Paid/Unpaid'].str.contains("No", na = False )])/len(df)

unpaid = round(unpaid*100, 2)

length = len(df)

print("Unpaid internships currently make up {}%".format(unpaid), "of internships taken by the Entering Class of 2018")

print("Currently, there are", length, " students who have completed the internship form")
```
